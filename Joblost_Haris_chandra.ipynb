```python
import pandas as pd

# Load the dataset
data = pd.read_csv('joblost.csv')
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
#Dropping Unneeded Columns
data = data.drop(columns=['max', 'tmax'])
```

```python
data['cbsa'].dtype
```

```python
#Transforming Data Types
data['cbsa'] = data['cbsa'].astype('str')
```

```python
data['cbsa'].dtype
```

```python
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
data = data.dropna(axis=0)
```

```python
#Renaming Columns
data = data.rename(columns={
    'total_workers_employed' : 'total_workers',
    'worker_job_loss_rate'   : 'worker_to_jobloss_rate'
})
```

```python
#Creating a Calculated Field
data['cbsa_cbsaName'] = data['cbsa'] + ', ' + data['cbsa_name']
```

```python
data.head()
```

```python
#Remove Duplicate Rows
data = data.drop_duplicates()
```

```python
data.to_csv('cleaned_data.csv', index=False)
```
